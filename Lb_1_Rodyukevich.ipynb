```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(r"image.jpg")
plt.imshow(img)
```

```python
import numpy as np

img_gray = np.mean(img, axis=-1)
plt.imshow(img_gray, cmap='gray')
```

```python
sobel_matrix_1 = np.array([[-1, 0 , 1],[-2, 0, 2],[-1, 0 , 1]])
sobel_matrix_2 = np.array([[-1, -2 , -1],[0, 0, 0],[1, 2 , 1]])
[x, y] = np.shape(img_gray)  
sob_image = np.zeros(shape=(x, y))
for i in range(x - 2):
    for j in range(y - 2):
        Gx = np.sum(np.multiply(sobel_matrix_1 , img_gray[i:i + 3, j:j + 3]))  
        Gy = np.sum(np.multiply(sobel_matrix_2 , img_gray[i:i + 3, j:j + 3]))  
        sob_image[i+1 , j+1 ] = np.sqrt(Gx ** 2 + Gy ** 2)
plt.imshow(sob_image, cmap = 'gray') 
```
